# Data Description
- Background: 流失資料集的測試檔案包含 64,374 筆客戶紀錄，作為一個獨立的資料集，用來評估已訓練完成的客戶流失預測模型在效能與泛化能力上的表現。內含有訓練集與測試集。
- date received: 20260112
- Path to data : https://www.kaggle.com/datasets/muhammadshahidazeem/customer-churn-dataset?utm_source=chatgpt.com
- Unit of observation: 每一筆紀錄都對應一位客戶，並且包含與訓練檔案相同的一組特徵。
- Sample period: 依資料內容顯示涵蓋 2013 年至 2019 年（不同動物的進所月份略有差異）。
- Known issues:
  - Gender、Subscription Type、Contract Length 為文字類別欄位，需轉換為數值編碼後才能用於模型訓練。
  - Last Interaction、Payment Delay、Support Calls 可能存在極端值（outliers），需檢查並進行適當處理。
  - Total Spend、Usage Frequency、Tenure 尺度差異大，建模前應進行標準化（scaling）。
  - Churn 為目標變數（標籤），需確認其編碼方式（例如 0/1 或 Yes/No）。

- Definition for each variable:
  - CustomerID：客戶的唯一識別碼，用來區分不同客戶。
  - Age：客戶年齡（數值型）。
  - Gender：客戶性別（類別型，Male、Female）。
  - Tenure：客戶與公司維持關係的時間長度（月）。
  - Usage Frequency：客戶使用服務的頻率（數值型），反映客戶活躍程度。
  - Support Calls：客戶在一定期間內聯絡客服的次數。
  - Payment Delay：客戶付款延遲的次數或天數。
  - Subscription Type：客戶的訂閱方案類型（類別型，例如 Basic、Standard、Premium）。
  - Contract Length：客戶的合約長度（例如 Monthly、Yearly），表示與公司的綁定程度。
  - Total Spend：客戶累積消費金額（數值型），代表客戶價值。
  - Last Interaction：距離客戶最近一次與公司互動的時間（例如天數）。
  - Churn：是否流失的標籤（目標變數，1 = 流失，0 = 未流失）。
